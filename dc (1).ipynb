```python
import pandas as pd
df=pd.read_csv("covid78.csv")
df
```

```python
df.drop(df.columns[[0,1,-1,-2,-3,-4,-5,-6,-9,-10,-11,12]],axis=1,inplace=True)
```

```python
df=df.iloc[:-1-2-3-4-5]
```

```python
df= df.fillna(str(0))
```

```python
df
```

```python
import pandas as pd
for i in df.columns:
    df[i]=df[i].apply(lambda x:x.replace(',','').replace('+',''))
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns=['CountryOther','TotalCase','NewCases','TotalDeaths','NewDeaths','TotalRecovered','NewRecovered','ActiveCases','SeriousCritical','TotCases1M pop','Population','Continent']
```

```python
df["TotalCase"]=df['TotalCase'].astype(float)
df["NewCases"]=df['NewCases'].astype(float)
df["TotalDeaths"]=df['TotalDeaths'].astype(float)
df["NewDeaths"]=df['NewDeaths'].astype(float)
df["TotalRecovered"]=df['TotalRecovered'].astype(float)
df["NewRecovered"]=df['NewRecovered'].astype(float)
df["ActiveCases"]=df['ActiveCases'].astype(float)
df["NewRecovered"]=df['NewRecovered'].astype(float)
df["SeriousCritical"]=df["SeriousCritical"].astype(float)
df["TotCases1M pop"]=df["TotCases1M pop"].astype(float)
df["Population"]=df["Population"].astype(float)
```
